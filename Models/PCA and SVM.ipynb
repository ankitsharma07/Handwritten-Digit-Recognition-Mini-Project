# Splitting the dataset for training and testing

# Model Training

```python
# Create a function for PCA and SVM

def n_component_analysis(n, X_train, y_train, X_val, y_val):
    start = time.time()
    pca = PCA(n_components=n)
    print("PCA analysis begins with components: ", n)
    
    pca.fit(X_train) #reduces dimensions
    X_train_pca = pca.transform(X_train)
    X_val_pca = pca.transform(X_val)
        
    print("Starting SVM")
    classifier = svm.SVC()
    classifier.fit(X_train_pca, y_train)
    
    accuracy = classifier.score(X_val_pca, y_val)
    
    end = time.time()
    
    print("accuracy: ", accuracy, "time elapsed: ", int(end-start))
```

```python
model_running = np.linspace(0.70, 0.85, num=15)
accuracy = []

for n in model_running:
    temp = n_component_analysis(n, X_train, y_train, X_val, y_val)
    accuracy.append(temp)
```
